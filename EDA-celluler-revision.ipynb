##### Gender

Value gender tidak terlalu berpengaruh kepada tingkat churn rate pada provider layanan.

##### SeniorCitizen

Dari visualisasi diatas terlihat bahwa hampir setengah dari pengguna layanan yang sudah berumur lanjut berpindah ke provider lain. Ini menunjukan kebanyakan churn user dari warga yang sudah berumur. Provider seluler bisa memperhatikan hal ini demi menjaga user yang telah lanjut usia

##### Dependents

Presentase berpindahnya pengguna layanan yang memiliki tanggungan 2x lebih besar perpindahannya dibanding yang tidak memiliki tanggungan. Marketing dari provider seluler bisa sekira nya lebih giat mencari tau kebutuhan pengguna layanan seluler yang memiliki tanggunan sehingga pengguna layanan yang memiliki tanggugan bisa tetap menjadi pengguna layanan seluler dan tidak pindah ke provider lain.

##### PaperlessBilling

Pengguna layanan dengan tagihan kertas memiliki tingkat perpindahan 2x lebih besar dari pengguna layanan dengan tagihan elektronik. Hal ini menunjukan bahwa layanan dengan tagihan elektronik memiliki sistem lebih baik sehingga bisa membuat pelanggan tetap bertahan dengan penyedia layanan seluler.

##### MultipleLines

Dari data diatas menunjukan bahwa pengguna layanan dengan lebih dari satu nomor memiliki churn user tertinggi dengan 28% dibandingan yang lainnya, namun tidak terlalu jauh dengan pengguna jasa tampa nomor lebih dari satu dan pengguna jasa layanan tampa telefon servis. Hal ini meunjukan bahwa ada nya sistem yang mungkin kurang disukai dari pengguna layanan lebih dari satu nomor.


##### Contract
Pada bagian Contract, dapat dilihat bahwa semakin lama jangka kontrak dengan pelanggan akan membuat churn rate menjadi lebih rendah. Kontrak per bulan menghasilkan churn rate 42,7% sedangkan kontrak 2 tahun hanya menghasilkan churn rate 2,8%. Dari hal ini bisa pihak provider seluler bisa mencari cara bagaimana meyakinkan pelanggan untuk mengambil kontrak 2.
